```
import pandas as pd
import numpy as np
import math
import re
```

```
l=['corona','coronavirusupdate','coronavirus','coronavius','ncov','coronavirius','chinavirus','corononavirus','covid','wuhancoronavirus','coronarovirus','novelcoronavirus','chinacoronavirus','coronarvirus','wuhanvirus','virus','coronarivus']
def top_10_hashtags(f,s):
  tdf=pd.read_csv(f)
  print(tdf.shape)
  tdf['hashtags']=tdf['hashtags'].str.replace("'",'')
  tdf['hashtags']=tdf['hashtags'].str.replace("[",'')
  tdf['hashtags']=tdf['hashtags'].str.replace("]",'')
  
  flist=[]
  for x in tdf['hashtags']:
    if(type(x)==type(float('nan'))):
      continue

    x=x.lower()
    y=x.split(',')
    for z in y:
      if(z!=''):
        z1=re.sub('[^a-zA-Z]','',z)
        if(z1!='' and z1 not in l):
          flist.append(z.strip())
  print(len(flist))
  counts=pd.DataFrame()
  counts['hashtags']=flist
  
  counts['hashtags']=counts[~counts['hashtags'].isin(l)] 
  y=counts['hashtags'].value_counts().nlargest(10)
  y.to_csv(s)
```
