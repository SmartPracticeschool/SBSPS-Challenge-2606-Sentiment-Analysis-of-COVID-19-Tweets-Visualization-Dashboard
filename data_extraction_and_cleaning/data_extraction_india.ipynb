Extracting Tweet IDs and Geolocation data from downloaded JSON files of tweets from Indian locations and saving them to text files

```
import json
import pandas as pd
import zipfile
import shutil
import os
import gc
```

```
def extract(fname,sname):
  print('here')
  for line in open(fname+'.json', 'r'):
    x=json.loads(line)
    if(len(x["geo"])!=0):
      if(x["geo"]["country_code"]=="in"):
        with open(sname+'.txt', "a+") as output:
          output.write(str(x["tweet_id"]+" , "))
          output.write(str(x["geo"]))
          output.write(str("\n"))
    elif(len(x["place"])!=0):
      if(x["place"]["country_code"]=="in"):
        with open(sname+'.txt', "a+") as output:
          output.write(str(x["tweet_id"]+" , "))
          output.write(str(x["place"]))
          output.write(str("\n"))
    elif(len(x["user_location"])!=0):
      if(x["user_location"]["country_code"]=="in"):
        with open(sname+'.txt', "a+") as output:
          output.write(str(x["tweet_id"]+" , "))
          output.write(str(x["user_location"]))
          output.write(str("\n"))
```

```
for i in range(1,30):
  if(i<10):
    j='0'+f'{i}'
  else:
    j=f'{i}'
  with zipfile.ZipFile("/content/drive/My Drive/IBM_sentiment_analysis/data/downloaded_geodata_json_files/geo_2020-02-"+j+".zip","r") as zip_ref:
    zip_ref.extractall()
  extract('geo_2020-02-'+j,'geo_2020-02-'+j)
  shutil.move('geo_2020-02-'+j+'.txt', '/content/drive/My Drive/IBM_sentiment_analysis/data/json_to_text(filtering_india)/geo_2020-02-'+j+'.txt')
  os.remove('geo_2020-02-'+j+'.json')
```
