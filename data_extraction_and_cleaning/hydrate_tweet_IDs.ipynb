Hydrating the Tweet IDs to obtain full tweet text and other required information and saving data to CSV files

Twarc is used for hydration of Tweet IDs

https://github.com/DocNow/twarc

```
def is_json_key_present(json, key):
    try:
        buf = json[key]
    except KeyError:
        return False
    return True
```

```
import csv
import json
import shutil
import os
for i in range(1,31):
  if(i<10):
    j='0'+f'{i}'
  else:
    j=f'{i}'
  txtfile=open('/content/drive/My Drive/IBM_sentiment_analysis/data/json_to_text(filtering_india)/geo_2020-04-'+j+'.txt')
  lines=txtfile.readlines()
  ids=[]
  for line in lines:
    ids.append(line.split(',')[0].strip())
  with open('text_geo_2020-04-'+j+'.txt', "w") as fout:
    print(*ids, sep="\n", file=fout)

  hydrate='twarc hydrate '+'text_geo_2020-04-'+j+'.txt'+'>'+'temp_geo_2020-04-'+j+'.json'
  !eval {hydrate}
  output=csv.writer(open('hy_geo_2020-04-'+j+'.csv','a+'))
  for tweetj in open('temp_geo_2020-04-'+j+'.json','r'):
    tweet=json.loads(tweetj)
    if(is_json_key_present(tweet, 'retweeted_status')==False):
      row=[]
      row.append(tweet['created_at'])
      row.append(tweet['id'])
      row.append(tweet['id_str'])
      row.append(tweet['full_text'])
      row.append(tweet['entities']['hashtags'])
      row.append(tweet['user']['location'])
      row.append(tweet['geo'])
      row.append(tweet['coordinates'])
      row.append(tweet['place'])
      row.append(tweet['retweet_count'])
      row.append(tweet['lang'])
    else:
      row=[]
      row.append(tweet['created_at'])
      row.append(tweet['id'])
      row.append(tweet['id_str'])
      row.append(tweet['retweeted_status']['full_text'])
      row.append(tweet['retweeted_status']['entities']['hashtags'])
      row.append(tweet['user']['location'])
      row.append(tweet['geo'])
      row.append(tweet['coordinates'])
      row.append(tweet['place'])
      row.append(tweet['retweet_count'])
      row.append(tweet['lang'])
    output.writerow(row)
  shutil.move('hy_geo_2020-04-'+j+'.csv', '/content/drive/My Drive/IBM_sentiment_analysis/data/hydrated_csv_files/hy_geo_2020-04-'+j+'.csv')
  os.remove('text_geo_2020-04-'+j+'.txt')
  os.remove('temp_geo_2020-04-'+j+'.json')
  os.remove('twarc.log')
```
