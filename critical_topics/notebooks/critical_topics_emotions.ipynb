```
import pandas as pd
import numpy as np
```

```
df=pd.read_csv('/content/drive/My Drive/IBM_sentiment_analysis/classification_sentiment_analysis/data/labelled_emotions(fear,anger,sadness,joy,neutral).csv')
print(df.shape)
print(df.isnull().sum())
df.head()
```

```
searchfor = ['worker','migrant','labourer','job']
```

```
df1=df[df['cleaned'].str.contains('|'.join(searchfor))]
print(df1.shape)
print(df1.isnull().sum())
df1.head()
```

```
df2=df1.iloc[:,1:]
df2=df2.reset_index(drop=True)
print(df2.shape)
df2.head()
```

```
len(df2['date'].unique())
```

```
import datetime
full=[]
summ=0
start_date = datetime.date(2020, 2, 1)
end_date = datetime.date(2020, 4, 30)
delta = datetime.timedelta(days=1)
while start_date <= end_date:
    temp=df2[df2['date']==str(start_date)]
    n=temp[temp['emotion']=='neutral']
    s=temp[temp['emotion']=='sadness']
    a=temp[temp['emotion']=='anger']
    j=temp[temp['emotion']=='joy']
    f=temp[temp['emotion']=='fear']
    nc=n.shape[0]
    sc=s.shape[0]
    ac=a.shape[0]
    jc=j.shape[0]
    fc=f.shape[0]
    total=nc+sc+ac+jc+fc
    if(total!=0):
      ncp=(nc/total)*100
      scp=(sc/total)*100
      acp=(ac/total)*100
      jcp=(jc/total)*100
      fcp=(fc/total)*100
    else:
      ncp=0
      scp=0
      acp=0
      jcp=0
      fcp=0

    row=[start_date,nc,sc,ac,jc,fc,total,ncp,scp,acp,jcp,fcp]
    full.append(row)
    summ=summ+nc+sc+ac+jc+fc
    start_date += delta
    print(start_date)
```

```
print(summ)
```

```
dff=pd.DataFrame(full,columns=['date','neutral_count','sadness_count','anger_count','joy_count','fear_count','total_count','neutral_count_pct','sadness_count_pct','anger_count_pct','joy_count_pct','fear_count_pct'])
print(dff.shape)
print(dff.isnull().sum())
dff.head()
```

```
dff.to_csv('/content/drive/My Drive/IBM_sentiment_analysis/critical_topics/emotion_plots/workers_emotion.csv',index=False)
```
