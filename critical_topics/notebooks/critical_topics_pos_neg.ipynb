```
df=pd.read_csv('/content/drive/My Drive/IBM_sentiment_analysis/data/labelled_positive_negative_neutral.csv')
print(df.shape)
print(df.isnull().sum())
df.head()
```

```
searchfor = ['economy','economycollapse','economy collapse','financialcrisis','financial crisis','debt']
```

```
df1=df[df['cleaned'].str.contains('|'.join(searchfor))]
print(df1.shape)
print(df1.isnull().sum())
df1.head()
```

```
df2=df1.reset_index(drop=True)
print(df2.shape)
df2.head()
```

```
len(df2['date'].unique())
```

```
import datetime
full=[]
summ=0
start_date = datetime.date(2020, 2, 1)
end_date = datetime.date(2020, 4, 30)
delta = datetime.timedelta(days=1)
while start_date <= end_date:
    temp=df2[df2['date']==str(start_date)]
    neu=temp[temp['sentiment']==0]
    pos=temp[temp['sentiment']==1]
    neg=temp[temp['sentiment']==-1]
    neu_c=neu.shape[0]
    pos_c=pos.shape[0]
    neg_c=neg.shape[0]
    total=neu_c+pos_c+neg_c
    if(total!=0):
      neu_cp=(neu_c/total)*100
      pos_cp=(pos_c/total)*100
      neg_cp=(neg_c/total)*100
    else:
      neu_cp=0
      pos_cp=0
      neg_cp=0

    row=[start_date,neu_c,pos_c,neg_c,total,neu_cp,pos_cp,neg_cp]
    full.append(row)
    summ=summ+neu_c+pos_c+neg_c
    start_date += delta
    print(start_date)
```

```
print(summ)
```

```
dff=pd.DataFrame(full,columns=['date','neutral_count','positive_count','negative_count','total','neutral_count_pct','positive_count_pct','negative_count_pct'])
print(dff.shape)
print(dff.isnull().sum())
dff.head()
```
