```
import pandas as pd
import numpy as np
```

```
df=pd.read_csv('/content/drive/My Drive/IBM_sentiment_analysis/data/labelled_positivity_score.csv')
print(df.shape)
print(df.isnull().sum())
df.head()
```

```
searchfor = ['pm care', 'pm care fund','care fund','pmcarefund','pmcare','carefund','pm cares fund','pmcaresfund','pm cares','cares fund','caresfund','pmcares','relief fund','relieffund','pmrelieffund','pm relief funds','pmrelieffunds','relieffunds']
```

```
df1=df[df['cleaned'].str.contains('|'.join(searchfor))]
print(df1.shape)
print(df1.isnull().sum())
df1.head()
```

```
df2=df1.reset_index(drop=True)
print(df2.shape)
df2.head()
```

```
len(df2['date'].unique())
```

```
import datetime
full=[]
summ=0
start_date = datetime.date(2020, 2, 1)
end_date = datetime.date(2020, 4, 30)
delta = datetime.timedelta(days=1)
while start_date <= end_date:
    temp=df2[df2['date']==str(start_date)]
    score=0
    count=0
    for index,row in temp.iterrows():
      score=score+row['sentiment']
      count=count+1
    if(count==0):
      row=[start_date,score,count,0]
    else:
      row=[start_date,score,count,(score/count)]
    full.append(row)
    summ=summ+count
    start_date += delta
    print(start_date)
```

```
print(summ)
```

```
dff=pd.DataFrame(full,columns=['date','total_score','count','normalised_score'])
print(dff.shape)
print(dff.isnull().sum())
dff.head()
```

```
dff.to_csv('/content/drive/My Drive/IBM_sentiment_analysis/critical_topics/score_plots/carefund_score.csv',index=False)
```
