```
def findmax(row):
  row=row.replace('[','')
  row=row.replace(']','')
  lt=row.split(',')
  lt=list(map(str.strip,lt))
  l=list(map(float,lt))
  m=max(l)
  i=l.index(m)
  l1=['joy', 'sadness', 'fear', 'anger', 'neutral']
  emotion=l1[i]
  f=[m*100,i,emotion]
  return f
```

```
f1['f']=f1['proba'].apply(findmax)
```

```
f1.head()
```

```
f1[['confidence','index','emotion']] = pd.DataFrame(f1.f.tolist(), index= f1.index)
```
